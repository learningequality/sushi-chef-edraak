```python
from pressurecooker.youtube import YouTubeResource
```

```python
user_url = 'https://www.youtube.com/user/CoursesTube'

yr = YouTubeResource(user_url)
```

```python
info = yr.get_resource_info()
```

```python
info.keys()
```

```python
videos = info['children']
len(videos)
```

```python
video = videos[3]
video
```

```python
from collections import defaultdict
licenses = defaultdict(int)
```

```python
for video in videos:
    licenses[video['license']] += 1
```
